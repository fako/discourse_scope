```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append('../')

import pandas as pd

from readers import JsonReader
from vectorizers import ArguingLexiconVectorizer
```

```python
json_reader = JsonReader(source="death_penalty.json", subjects=["death penalty", "capital punishment"])
texts = json_reader.get_texts()
```

```python
vec = ArguingLexiconVectorizer(stop_words="english")
vec.fit(texts)
feature_names = vec.get_feature_names()
vectors = vec.transform(texts)
frame = pd.DataFrame(vectors.toarray(), columns=feature_names)
```

```python
len(vec.get_feature_names())
```

```python
number_features = [feature for feature in feature_names if not feature.isalpha()]
frame.drop(labels=number_features + ['death', 'penalty', 'punishment'], axis=1, inplace=True)
words_sorted = frame.sum(axis=0).sort_values(ascending=False)
```
